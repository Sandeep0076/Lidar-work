## Shape, size, Dimention, range, arange, reshape, max,slice, index,ravel,size,

```python
f = nmpy.array([(1,2, 3), (11, 9, 7), (4, 5, 6),(44, 52, 62)])

# give shape ex (4,3)
g = f.shape
print(g)
# shape[0] = means how many rows = 4
#and shape[1] = how many columns = 3
h = f.shape[1]
# shows the dimention
i = f.ndim
print(i)
# size will show u total no. of elements
i = f.size
print(i)
# makes array of ones in shape of 4 by 3
i = nmpy.ones((4,3))
# range from 0 till 1o
list(range(10))

# make an array, Start, Stop,interval
i = list(nmpy.arange(5,50,2))
# same as range, start, stop and the last means total no. of values to be displayed with the equal distance to each space. 
i= nmpy.linspace(5,50,4)
#reshape((-1,4) means it moulds into four columns and according to that remaining rows will be decided automatically,
#-1 means unknown
i = nmpy.random.randint(5,150,21).reshape(-1,3)
# arguments, start,stop,size of matrix, and according to that no, of random int will be generated.
i = nmpy.random.randint(1,21,(10,4))
# sum of all the columns
j= i.sum(axis=0)
#axis=1 sum of all the rows
# maximum of 
nmpy.max(i)
# for taking out specific columns , arg= : , means to and from which rows and (0,2,3) means column no, 0,2 and 3
j = i[:2,(0,2,3)]
# access data at particula index
#type(i)
#change back to the list
k= i.ravel()
#knowing the size of the array
i.size
```

# Matplotlib basics

```python
import matplotlib.pyplot as plt 
import mpl_toolkits.mplot3d as axes3d
```

```python
x = nmpy.random.randint(1,5,10)
y = nmpy.random.randint(10,50,10)
z = nmpy.random.randint(30,50,10)
p = nmpy.random.randint(1,5,10)
q = nmpy.random.randint(1,5,10)
```

### For plotting 2d data

```python
#imput the parameters into it
#plt.plot(x,y) or plt.scatter(x,y)

# for labelng the axis
plt.xlabel('Xaxis')
plt.ylabel('Yaxis')
# for restrictinmg the axis , how much data should be shown : lower and upper limit for x and then lower and upper limit for Y
plt.axis([10,20, 1000,10000])

#show the ghraph
plt.show()
```

```python
# figure means area in total within which all the charts/fig1,fig2
#inside argument is figure no, and size
figure1 = plt.figure(1,figsize=(6,4))
#add subplot is used to show where in figure which row and column u want to display the chart, 111 means 1 fig, 1 row, 1 column

ax1 = figure1.add_subplot(121)
ax2=figure1.add_subplot(122)
ax1.plot(x,y)
ax2.plot(p,q)
ax1.set(xlabel='Xaxis',ylabel='yaxis')
plt.show()
```
